The data is clean with no missing or duplicated values and no outliers. This data does not require any preprocessing. So, We will now proceed with separating out Independent and Dependent variables and splitting the data into train and test.

```python
df.drop(columns = ['Serial No.'])  #dropping unnecessary column
```

```python
X = df.iloc[:,0:8]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
r2
```

```python
X_test.shape
```

```python
#Adjusted R2 score
1-(((1-r2)*(100-1))/(100-1-1))
```

The more closer the value of R2 is towards 1, the better the model is. Since the fitted linear regression model has a value of 0.79 I will consider it a good fit model. But I will be exploring other models on this data as I learn to see how score changes with them.
